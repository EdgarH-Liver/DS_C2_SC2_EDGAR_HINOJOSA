```python
#from sqlite3 import PARSE_DECLTYPES
# Reto de pandas
import pandas as pd

file = 'country_vaccinations.csv'

# 1. leer documento y convertir date
df_temp = pd.read_csv(f'{file}', sep = ',', encoding = 'ISO-8859-1', low_memory = False, parse_dates = ['date'])

# 2. Tipo de datos
# atributos_types = df_temp.dtypes # cantidad de registros y columnas
atributos_info = df_temp.info() # estructura y tipo de datos

print("Los tipos de datos por columna son: ", atributos_info)

# 3. Cantidad de vacunas aplicadas por lote

unicos_vaccines = df_temp['vaccines'].drop_duplicates().to_list() # lista de vacunas unicas

for vacuna in unicos_vaccines:
  vc = df_temp[df_temp['vaccines'].str.contains(vacuna)]['vaccines'].astype(str)

  size = len(vc) # cantidad de vacunas por paquetes de vacunas
  print('Vacuna y cantidad de vacuna: ', vacuna, ' -> ', size)

# 4. Total de vacunas en todo el mundo
total_vacunas = df_temp['daily_vaccinations'].sum()
print('Total de vacunas aplicadas en el mundo: ', total_vacunas )

# 5 Promedio de vacunas por pais
paises = df_temp['country'].drop_duplicates().to_list() # lista de paises unicos
for pais in paises:
  df_pais = df_temp[df_temp['country'] == pais]
  vacunas = df_pais['daily_vaccinations'].astype(float).sum()
  size = len(df_pais)
  promedio = vacunas/size
  print('Promedio de vacunas en ', pais, ' es de: ', promedio)

# paises = df_temp['country'].drop_duplicates().count()
# promedio = total_vacunas/paises
# print('Promedio de vacunas por pais: ', promedio)


#6. vacunas el 29/01/21

fecha_busqueda = pd.to_datetime('29/01/2021')
vacunas_dia = df_temp[df_temp['date']==fecha_busqueda]['daily_vaccinations'].sum()
print('Vacunas al ', fecha_busqueda, ': ', vacunas_dia)

# 7 nuevo data frame con diferencias

df_conDiferencias = df_temp.copy()
df_conDiferencias['Diferencias'] = df_conDiferencias['daily_vaccinations'] - df_conDiferencias['daily_vaccinations_raw']
# print(df_conDiferencias['Diferencias'])

# 8 perido de tiempo mas antiguo al reciente
fecha_min = df_temp['date'].min()
fecha_max = df_temp['date'].max()
periodo = fecha_max-fecha_min
print('Periodo de tiempo de vacunacion: ', periodo)

# 9 nuevo data frame con cantidad
df_conCantidad = df_temp.copy()
df_conCantidad['canVac'] = df_conCantidad.apply(lambda row: len(str(row['vaccines']).split(',')), axis=1)

# 10 data frame registros antes del 20 de diciembre de 2020.
fecha_antes = pd.to_datetime('20/12/2020')
df_antes20 = df_temp[df_temp['date']<fecha_antes]

# 11 dataframe pfizer
df_pfizer = df_temp[df_temp['vaccines'].str.contains('Pfizer')]
print('cantidad de registros con pfizer: ', len(df_pfizer))

# 12 guardar dataframe
with pd.ExcelWriter(f'resultadosReto.xlsx', engine = 'openpyxl') as writer:
            df_conDiferencias.to_excel(excel_writer=writer, sheet_name = 'ConDiferencias', index = False,
                            header = True, startrow = 1, startcol = 1)
            df_conCantidad.to_excel(excel_writer=writer, sheet_name = 'ConCantidad', index = False,
                            header = True, startrow = 1, startcol = 1)
            df_antes20.to_excel(excel_writer=writer, sheet_name = 'Antes20', index = False,
                            header = True, startrow = 1, startcol = 1)
            df_pfizer.to_excel(excel_writer=writer, sheet_name = 'Pfizer', index = False,
                            header = True, startrow = 1, startcol = 1)

```
